```python
import pandas as pd
import random as rdm
pd.set_option('display.max_columns', None)
```

```python
#reads data from csv downloaded from basketballreference.com
#Data has been manually labeled in the "Better" column
pl = pd.read_csv("NBA 2021 Per 100.csv", header = 4)
pl = pl.drop(columns = ['Unnamed: 22'])
pl
```

```python
#Separate good players into a (gp) dataframe
gp = pl.loc[pl['Better'] == 1].copy()
gp
```

```python
#Same thing with bad players (bp)
bp = pl.loc[pl['Better'] == 0].copy()
bp
```

```python
#Create a dataframe with matchups featuring a good player and a bad player
#Left side takes from gp dataframe, 2400 entries from 24 random samples of 100 players
matchl = pd.DataFrame(columns = pl.columns)
for i in range(24):
    matchl = matchl.append(gp.sample(n=100), ignore_index=True).copy()
#matchl
```

```python
#right side from bp dataframe, 2400 entries from 6 random samples of 400 players
matchr = pd.DataFrame(columns = pl.columns)
for i in range(6):
    matchr = matchr.append(bp.sample(n=400), ignore_index=True).copy()
#matchr
```

```python
#randomize the lists so that the both sides contain both good and bad players
for i in range(len(matchl.index)):
    coin = rdm.randint(0,1)
    if coin == 1:
        temp1, temp2 = matchl.iloc[i].copy(), matchr.iloc[i].copy()
        matchl.iloc[i], matchr.iloc[i] = temp2, temp1
matchl = matchl.drop(columns = ['Better'])
```

```python
#conccatenate the two sides together
matchup = pd.concat([matchl, matchr], axis = 1).copy()
matchup = matchup.drop_duplicates(ignore_index = True)
#pd.set_option('display.max_columns', None)
matchup
```

```python
#separate the label from the data
data = matchup.drop(columns = ['Player','Better','ORtg','DRtg']).copy()
label = matchup['Better'].copy()
```

```python
#Assign a value based on role, from 0 for point guard (PG) to 1 for center (C)
datan = data.copy()
datan['Pos'] = datan['Pos'].replace({'PG':0, 'PG-SG':0.125, 'SG-PG':0.125, 'SG':0.25, 'SG-SF':0.375, 'SF-SG':0.375, 'SF':0.5, 
                                       'SF-PF':0.625, 'PF-SF':0.625, 'PF':0.75, 'PF-C':0.875, 'C-PF':0.875, 'C':1})
#normalize data
for i in datan.columns:
    #if tdatan[i].max() == 1 and tdatan[i].min() == 0:
    datan[i] = (datan[i] - datan[i].min())/datan[i].max()
#datan['ORtg'] = (datan['ORtg'] - datan['ORtg'].min())/(datan['ORtg'].max() - datan['ORtg'].min())
#pd.set_option('display.max_rows', None)
datan
```
